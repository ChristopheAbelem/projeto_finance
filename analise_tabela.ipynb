## 3. Analise de Precos 

- Por se tratar de um dataset financeiro, o mesmo contem um grande volume de dados relacionados a precos. Por isso, essa secao sera dividida em analisar o preco das comodites(que tambem estao subdivididas), das criptomoedas e do valor de mercado das empresas presentes no arquivo

- Ao se lidar com esse tipo de dados e seu volume, foi necessario analisar os preco de ano em ano para evitar sobrecarregamento e erros 

### 3.1. Precos de comodites 

```python
# Verificar as colunas que contêm 'Price' no nome
commodities_columns = [col for col in df.columns if 'Price' in col]

# Exibir a lista de colunas relacionadas às commodities
commodities_columns

# Criar um DataFrame com todas as colunas de preços
df_commodities = df[['Date'] + commodities_columns].copy()

# Converter a coluna de datas para o formato datetime
df_commodities['Date'] = pd.to_datetime(df_commodities['Date'], errors='coerce')

# Exibir as primeiras linhas do DataFrame
df_commodities.head()

```

```python
# Filtrar as commodities de metais preciosos
commodities_metais = ['Gold_Price', 'Silver_Price', 'Copper_Price']

# Criar um dataframe específico para os metais preciosos e converter a coluna 'Date' para datetime
df_metais = df[['Date'] + commodities_metais].copy()
df_metais['Date'] = pd.to_datetime(df_metais['Date'], errors='coerce')

# Remover valores inválidos (NaN)
df_metais.dropna(subset=['Date'], inplace=True)
df_metais[commodities_metais] = df_metais[commodities_metais].apply(pd.to_numeric, errors='coerce')

# Agrupar os dados por ano e tirar a média dos preços dos metais por ano
df_metais_resampled = df_metais.set_index('Date').resample('Y').mean().reset_index()

# Plotar os preços normalizados dos metais preciosos
plt.figure(figsize=(12, 6))

for commodity in commodities_metais:
    plt.plot(df_metais_resampled['Date'], df_metais_resampled[commodity], label=commodity)

plt.title('Preços Médios Anuais dos Metais Preciosos (Ouro, Prata, Cobre)')
plt.xlabel('Data')
plt.ylabel('Preço Médio')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Selecionar as commodities de metais preciosos
commodities_metais= ['Date', 'Gold_Price', 'Silver_Price', 'Copper_Price']

# Criar o DataFrame de metais preciosos
df_metais = df[commodities_metais].copy()

# Verificar as primeiras linhas
df_metais.head()
```
